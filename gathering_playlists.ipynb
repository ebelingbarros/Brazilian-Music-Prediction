# Extracting musics from Spotify

-------

### Importing libraries

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import getpass
```

```python
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id="df77ca1e056a425eaaca06513b26fa41"                                                          
                                                           , client_secret="cb2222815b8a40afa8475581fe8c7081"
                                                          ))
```


### Importing sertanejo music playlist

```python
sertanejo = sp.user_playlist_tracks('spotify', '37i9dQZF1DX5lAV5ZGu9ik')
```

```python
sertanejo
```

```python
len(sertanejo['items'])
```

```python
sertanejo['items'][0]['track']['id']

for item in sertanejo['items']:
    print(item['track']['id'])

sertanejo1 = []
for item in sertanejo['items']:
    sertanejo1.append(item['track']['id'])
```

```python
sertanejo1
```

```python
sp.audio_features(tracks=sertanejo1)
```

```python
import pandas as pd
```

```python
audio_features_sertanejo1 = pd.DataFrame(sp.audio_features(tracks=sertanejo1))
```

```python
audio_features_sertanejo1
```

#### Exporting the extracted playlist

```python
audio_features_sertanejo1.to_csv('sertanejo.csv', index = True)
```
