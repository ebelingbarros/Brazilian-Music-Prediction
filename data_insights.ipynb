## Gathering insights through visualizations

```python
import pandas as pd
```

```python
data = pd.read_csv('insights.csv')
data.head()
```

### Plotting track features

```python
import matplotlib.pyplot as plt
fig,a =  plt.subplots(2,2, figsize=(8,8))
x = data['track_no']
a[0][0].plot(x,data['danceability'])
a[0][0].set_title('danceability')
a[0][1].plot(x,data['energy'])
a[0][1].set_title('energy')
a[1][0].plot(x,data['loudness'])
a[1][0].set_title('loudness')
a[1][1].plot(x,data['valence'])
a[1][1].set_title('valence')
plt.show()
```

### Cluster distribution histogram

```python
data['cluster_no'].plot(kind='hist')
```

### Cluster scatter plot

```python
# Extract the data we're interested in
danceability, energy = data['danceability'], data['energy']

# Scatter the points, using size and color but no label
plt.scatter(danceability, energy, label=None,
                   c=data['cluster_no'], cmap='viridis',
                   )
plt.xlabel('danceability')
plt.ylabel('energy')
plt.colorbar(label='cluster number')
```
